```python
import pandas as pd

# Load the dataset
file_path = '/content/Covid data.csv'  # Adjust path as necessary
df = pd.read_csv(file_path)

# Convert 'date' column to datetime
df['date'] = pd.to_datetime(df['date'], format='%d/%m/%Y')

# Fill missing values (you might choose different strategies based on your analysis)
df.fillna(0, inplace=True)

# Check for duplicate entries
df.drop_duplicates(inplace=True)

# Display cleaned data
print(df.head())
```

Clean Up Numeric Columns

 we'll focus on making sure numeric columns are in the correct format. For instance, the total_vaccinations, daily_vaccinations, and other numeric columns should not have any non-numeric values.

```python
# Convert relevant columns to numeric, coercing errors to NaN
df['total_vaccinations'] = pd.to_numeric(df['total_vaccinations'], errors='coerce')
df['people_vaccinated'] = pd.to_numeric(df['people_vaccinated'], errors='coerce')
df['daily_vaccinations'] = pd.to_numeric(df['daily_vaccinations'], errors='coerce')

# Check if there are any NaNs in these columns after conversion
print(df[['total_vaccinations', 'people_vaccinated', 'daily_vaccinations']].isnull().sum())
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

# Load the data
df = pd.read_csv('/content/Covid data.csv')

# Clean the data (this step will depend on your specific dataset)
# Example: Dropping rows with missing values in total_vaccinations
df = df.dropna(subset=['total_vaccinations'])

# Select features and target
features = df[['country']]  # Adjust as necessary
target = df['total_vaccinations']

# Split the data
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Encode categorical variables
X_train_encoded = pd.get_dummies(X_train, drop_first=True)
X_test_encoded = pd.get_dummies(X_test, drop_first=True)

# Align the training and test data
X_train_encoded, X_test_encoded = X_train_encoded.align(X_test_encoded, join='left', axis=1, fill_value=0)

# Fit the model
model = RandomForestRegressor()
model.fit(X_train_encoded, y_train)

# Make predictions
y_pred = model.predict(X_test_encoded)

# Evaluate the model
mae = mean_absolute_error(y_test, y_pred)
print(f'Mean Absolute Error: {mae}')
```
